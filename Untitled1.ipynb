```python
import pandas as pd

io_params = {
    'parse_dates': ['epoch']
}

train = pd.read_csv('data/IDAO 2020/train.csv', **io_params)
```

Remove duplicate entries.

```python
import datetime as dt
import tqdm

train['epoch_diff'] = train.groupby('sat_id')['epoch'].diff()
dtypes = train.dtypes.to_dict()

cols_to_shift = train.columns.difference(['x_sim', 'y_sim', 'z_sim', 'Vx_sim', 'Vy_sim', 'Vz_sim'])

train_sats = []

for sat_id in tqdm.tqdm(train['sat_id'].unique()):
    
    g = train.query('sat_id == @sat_id').copy()
    dups = g[g['epoch_diff'] < dt.timedelta(seconds=60)].index
    
    for i in reversed(dups):
        g.loc[i:, cols_to_shift] = g.loc[i:, cols_to_shift].shift(-1)
        
    g = g.drop(g[g['x'].isnull()].index)
        
    train_sats.append(g)
    
train = pd.concat(train_sats).astype(dtypes).drop(columns='epoch_diff')
```

Merge train and test.

```python
test = pd.read_csv('data/IDAO 2020/Track 1/test.csv', **io_params)
data = pd.concat((train, test), sort=False)
data['is_train'] = data['x'].notnull()
data = data.sort_values(['sat_id', 'epoch'])
data['is_test_sat'] = data['sat_id'].isin(data.query('not is_train')['sat_id'].unique())
```

Add cycle information.

```python
data['cycle_index'] = data.groupby('sat_id').apply(lambda g: pd.Series(pd.np.arange(len(g)) % 24)).to_numpy()
data['cycle_no'] = data.groupby('sat_id').apply(lambda g: pd.Series(pd.np.arange(len(g)) // 24)).to_numpy()
```

Implement SMAPE.

```python
import numpy as np

def smape(y_true, y_pred): 
    return np.mean(np.abs((y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true))))
```

```python
import functools
import tqdm

preds = []


for (sat, idx), g in tqdm.tqdm(data.groupby(['sat_id', 'cycle_index']), mininterval=1):
    
    train = g[g['is_train']]
    cut_point = len(train) // 2 + 1
    fit = train[:cut_point]
    val = train[cut_point:]
    test = g[~g['is_train']]
    
    for col in ('x', 'y', 'z', 'Vx', 'Vy', 'Vz'):
        
        x_fit = fit['epoch'].astype(pd.np.int64) // 10e9 / 10e6
        y_fit = fit[col] / fit[f'{col}_sim']
        x_val = val['epoch'].astype(pd.np.int64) // 10e9 / 10e6
        
        poly = pd.np.polyfit(x_fit, y_fit, deg=2)
        pred = functools.reduce(np.add, (coef * x_val ** i for i, coef in enumerate(reversed(poly))))
        
        preds.append(pd.DataFrame({
            'sat_id': sat,
            'epoch': val['epoch'],
            'cycle_index': idx,
            'y_true': val[col],
            'y_pred': pred * val[f'{col}_sim'],
            'target': col
        }))
        
preds = pd.concat(preds)
```

```python
sat_smapes = preds.groupby('sat_id').apply(lambda g: smape(g['y_true'], g['y_pred']))
mean_smape = sat_smapes.mean()
100 * (1 - mean_smape)
```
